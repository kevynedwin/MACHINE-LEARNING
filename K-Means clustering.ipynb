```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#import missingno as msno
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs #Make artificial data
```

```python
data=make_blobs( n_samples=200,
                 n_features=2,
                 centers=4,
                 cluster_std=1.8,
                 random_state=101) #Create Data
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans #create cluster 
```

```python
#k cluster
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0]) #fitting
```

```python
kmeans.cluster_centers_ #get the cluster centres
```

```python
kmeans.labels_
```

```python
f,(ax1,ax2)=plt.subplots(1,2,figsize=(10,6))

ax2.set_title('K Means')
ax2.scatter(data[0][:,0],data[0][:,1],c= kmeans.labels_,cmap='rainbow')

ax1.set_title('Original')
ax1.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
